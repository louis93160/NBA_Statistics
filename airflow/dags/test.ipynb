```python
import datetime
from datetime import datetime, date

df = pd.read_csv('test_points.csv')
points_df = df.rename(
    columns={
        "Team": "team",
        "Date": "game_date",
        "Opp": "opponent",
        "Result": "team_result",
        "FG": "team_field_goals",
        "FGA": "team_field_goals_attempted",
        "FG%": "team_field_goals_pctg",
        "2P": "team_2pts",
        "2PA": "team_2pts_attempted",
        "2P%": "team_2pts_pctg",
        "3P": "team_3pts",
        "3PA": "team_3pts_attempted",
        "3P%": "team_3pts_pctg",
        "FT": "team_fts",
        "FTA": "team_fts_attempted",
        "FT%": "team_fts_pctg",
        "PTS.1": "team_points",
        "FG.1": "opp_field_goals",
        "FGA.1": "opp_field_goals_attempted",
        "FG%.1": "opp_field_goals_pctg",
        "2P.1": "opp_2pts",
        "2PA.1": "opp_2pts_attempted",
        "2P%.1": "opp_2pts_pctg",
        "3P.1": "opp_3pts",
        "3PA.1": "opp_3pts_attempted",
        "3P%.1": "opp_3pts_pctg",
        "FT.1": "opp_fts",
        "FTA.1": "opp_fts_attempted",
        "FT%.1": "opp_fts_pctg",
        "PTS.2": "opp_points"
    }
).drop(columns=["Rk", "PTS", "MP"])

df1 = pd.read_csv('test_rebounds.csv')
rebounds_df = df1.rename(
    columns={
        "Team": "team",
        "Date": "game_date",
        "Opp": "opponent",
        "Result": "team_result",
        "ORB": "team_off_rebounds",
        "DRB": "team_def_rebounds",
        "TRB.1": "team_total_rebounds",
        "ORB.1": "opp_off_rebounds",
        "DRB.1": "opp_def_rebounds",
        "TRB.2": "opp_total_rebounds"
    }
).drop(columns=["Rk", "TRB", "MP"])


df2 = pd.read_csv('test_assists.csv')
assists_df = df2.rename(
    columns={
        "Team": "team",
        "Date": "game_date",
        "Opp": "opponent",
        "Result": "team_result",
        "AST.1": "team_assists",
        "STL": "team_steals",
        "BLK": "team_blocks",
        "TOV": "team_turnovers",
        "PF": "team_personal_fouls",
        "AST.2": "opp_assists",
        "STL.1": "opp_steals",
        "BLK.1": "opp_blocks",
        "TOV.1": "opp_turnovers",
        "PF.1": "opp_personal_fouls"
    }
).drop(columns=["Rk", "AST", "MP"])

merged_df = pd.merge(points_df, rebounds_df, on=['team','game_date', 'opponent', 'team_result'])
final_df = pd.merge(merged_df, assists_df, on=['team','game_date', 'opponent', 'team_result'])
final_df["created_at"] = datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S')
# final_df["created_at"] = datetime.today().strftime('%Y-%m-%d %H:%M:%S')

final_df.to_csv('../data/game_log_2023-11-05.csv', index=False)
```
